```python
import requests
from bs4 import BeautifulSoup
import re
from colorama import init, Fore, Style

# Inisialisasi Colorama
init(autoreset=True)

print(Fore.MAGENTA + """
 ██████╗ ██████╗ ██╗   ██╗ ██████╗  ██████╗ ██████╗ ███╗   ███╗
██╔════╝██╔═══██╗██║   ██║██╔════╝ ██╔═══██╗██╔══██╗████╗ ████║
██║     ██║   ██║██║   ██║██║  ███╗██║   ██║██████╔╝██╔████╔██║
██║     ██║   ██║██║   ██║██║   ██║██║   ██║██╔═══╝ ██║╚██╔╝██║
╚██████╗╚██████╔╝╚██████╔╝╚██████╔╝╚██████╔╝██║     ██║ ╚═╝ ██║
 ╚═════╝ ╚═════╝  ╚═════╝  ╚═════╝  ╚═════╝ ╚═╝     ╚═╝     ╚═╝
""")
print(Fore.CYAN + "Cyb00M - Detekai URL/Link")
print(Style.BRIGHT + Fore.GREEN + "============================================================")


# Blacklist domain dan kata kunci mencurigakan
blacklist_domains = ["paypal-security.com", "secure-facebook-login.com"]
suspicious_keywords = ["password", "secure", "login", "verify"]

# Fungsi utama
def check_url(url):
    # Validasi format URL
    if not re.match(r'https?://', url):
        return "URL tidak valid. Pastikan dimulai dengan 'http://' atau 'https://'."

    # Periksa HTTPS
    if not url.startswith("https://"):
        return "URL tidak aman (tidak menggunakan HTTPS)."

    # Periksa blacklist domain
    blacklist_result = check_blacklist(url)
    if blacklist_result:
        return blacklist_result

    # Periksa struktur URL
    structure_result = check_url_structure(url)
    if structure_result:
        return structure_result

    # Periksa pengalihan
    redirect_result = check_redirect_chain(url)
    if redirect_result:
        return redirect_result  # Kembalikan hasil pengalihan

    # Analisis konten halaman
    content_result = analyze_content_simulation(url)
    if content_result:
        return content_result

    return "URL tidak mencurigakan berdasarkan pemeriksaan lokal. Tetap waspada!"

# Fungsi cek blacklist
def check_blacklist(url):
    for domain in blacklist_domains:
        if domain in url:
            return f"URL mencurigakan! Ditemukan domain dalam blacklist: {domain}"
    return None

# Fungsi cek struktur URL
def check_url_structure(url):
    if len(url.split(".")) > 3:
        return "URL mencurigakan! Terlalu banyak subdomain."
    if any(char in url for char in ["@", "%", "$", "&", "!", "~"]):
        return "URL mencurigakan! Mengandung karakter tidak biasa."
    return None

# Fungsi cek pengalihan
def check_redirect_chain(url):
    try:
        response = requests.get(url, allow_redirects=True)
        if len(response.history) > 0:
            return f"URL memiliki pengalihan ke: {response.url}"
    except Exception as e:
        return f"Kesalahan saat memeriksa redirect: {e}"
    return None

# Fungsi analisis konten simulasi
def analyze_content_simulation(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Cek form
        forms = soup.find_all('form')
        if forms:
            return "Halaman memiliki form. Periksa apakah meminta data sensitif."

        # Cek kata kunci mencurigakan
        for keyword in suspicious_keywords:
            if keyword in response.text.lower():
                return f"Halaman mencurigakan! Ditemukan kata kunci: {keyword}"
    except Exception as e:
        return f"Kesalahan saat menganalisis konten: {e}"
    return None

# Penggunaan script
if __name__ == "__main__":
    url = input("Masukkan URL yang ingin diperiksa: ")
    result = check_url(url)
    print(result)
```
